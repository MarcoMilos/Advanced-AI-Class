```python
import seaborn as sns
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Missing Values
print(titanic.isnull().sum())
```

```python
data = titanic[['age', 'fare', 'pclass', 'sex']]
data = pd.get_dummies(data, drop_first=True) # Convierte la variable 'sexo' en una columna binaria
data.head()
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0)
```

```python
data_imputed = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
```

```python
print("Datos originales")
print(data.head(30))
print("Datos imputados")
print(data_imputed.head(30))
```

## Ejercicio 1:

Use MICE on the life-expectancy dataset https://www.kaggle.com/datasets/kumarajarshi/life-expectancy-who

```python
# Authenticate and mount Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
import kagglehub
import os
import shutil
import pandas as pd

# Download latest version of the dataset
path = kagglehub.dataset_download("kumarajarshi/life-expectancy-who")

print("Path to dataset files:", path)

# Define the destination path in Google Drive
drive_path = "/content/drive/MyDrive/AI/Files/"

# Create the directory in Drive if it doesn't exist
os.makedirs(drive_path, exist_ok=True)

# Copy the downloaded files to the specified path in Google Drive
for item in os.listdir(path):
    s = os.path.join(path, item)
    d = os.path.join(drive_path, item)
    if os.path.isdir(s):
        # Use copytree for directories
        shutil.copytree(s, d, dirs_exist_ok=True)
    else:
        # Use copy2 for files
        shutil.copy2(s, d)

print(f"Dataset copied to: {drive_path}")
```

```python
# Define the path to the dataset in Google Drive
life_expectancy_path = "/content/drive/MyDrive/AI/Files/Life Expectancy Data.csv"

# Load the dataset
import pandas as pd # Import pandas here
life_expectancy_df = pd.read_csv(life_expectancy_path)

# Display the first few rows
print("Primeras filas del dataset de Life Expectancy:")
display(life_expectancy_df.head(10))

# Display the summary of missing values
print("\nResumen de valores faltantes antes de la imputación:")
print(life_expectancy_df.isnull().sum())
```

```python
# Select a subset of numerical columns for imputation
selected_cols = ['Life expectancy ', 'Adult Mortality', 'Alcohol', 'Hepatitis B', 'GDP']
life_expectancy_subset = life_expectancy_df[selected_cols]

# Initialize the IterativeImputer
imputer = IterativeImputer(max_iter=10, random_state=0)

# Apply MICE imputation
life_expectancy_imputed_subset = pd.DataFrame(imputer.fit_transform(life_expectancy_subset), columns=selected_cols)

# Display the imputed data
print("\nDatos de Life Expectancy después de la imputación con MICE (columnas seleccionadas):")
display(life_expectancy_imputed_subset.head(10))

# Final comparison between the two versions
print("\nDatos originales:")
print(life_expectancy_subset.head(30))
print("\nDatos imputados:")
print(life_expectancy_imputed_subset.head(30))
```

## Ejercicio 2:

Use MICE on the Planets (seaborn) dataset to fill in missing values.

```python
import seaborn as sns
import pandas as pd
import os
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

planets = sns.load_dataset('planets')
planets.head()
```

```python
# Missing values
print(planets.isnull().sum())
```

```python
planetsDT = planets[['orbital_period', 'mass', 'distance']]
planetsDT = pd.get_dummies(planetsDT, drop_first=True) # Convierte la variable 'sexo' en una columna binaria
planetsDT.head()
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0)
```

```python
planets_imputed = pd.DataFrame(imputer.fit_transform(planetsDT), columns=planetsDT.columns)
```

```python
print("Datos originales")
print(planetsDT.head(30))
print("Datos imputados")
print(planets_imputed.head(30))
```
