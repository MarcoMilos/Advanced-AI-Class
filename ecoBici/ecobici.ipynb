```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io

print("Librerias importadas correctamente")

# ==============================================================================================
```

```python
# leer el csv
csv_fle = pd.read_csv('./2025-04.csv')

# Mostrar el tama√±o del df (filas y columnas)
print("Tama√±o del df:")
display(csv_fle.shape)

# Mostrar una previsualizacion de los datos
print("Previsualizacion del df:\n")
display(csv_fle.head(50))
```

```python
# Regresar valores √∫nicos
print(csv_fle["Genero_Usuario"].unique())
```

```python
# Contamos cantidad de registros para cada genero
print(csv_fle["Genero_Usuario"].value_counts(dropna=False))
```

```python
# Agregamos todos los valores diferentes de "M" y "F" a "O" (other)
csv_fle["Genero_Usuario"] = csv_fle["Genero_Usuario"].where(csv_fle["Genero_Usuario"].isin(["M", "F"]), "O")
print(csv_fle["Genero_Usuario"].value_counts())
```

```python
# AHora vamos a cambiar los valores de edades de elementos flotantes a enteros (no necesitamos flotantes porque son edades)
csv_fle["Edad_Usuario"] = csv_fle["Edad_Usuario"].astype("Int64")
display(csv_fle.head(20))
```

```python
# Cambiar el nombre de la columna "Bici" a "Bici_UUID"
csv_fle.rename(columns={"Bici": "Bici_UUID"}, inplace=True)
display(csv_fle.head(20))
```
