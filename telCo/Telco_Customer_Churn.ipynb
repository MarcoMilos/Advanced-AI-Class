```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Telco_Customer_Churn.csv')
```

```python
df.head()
```

```python
plt.hist(df['tenure'], bins='auto')
```

```python
plt.hist(df['MonthlyCharges'], bins='auto')
```

```python
print('Datos estadísticos para tenure')
print('la media es',df['tenure'].mean())
print('la mediana es',df['tenure'].median())
```

```python
print('Datos estadísticos para MonthlyCharges')
print('la media es',df['MonthlyCharges'].mean())
print('la mediana es',df['MonthlyCharges'].median())
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ',0)
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
print('Datos estadísticos para TotalCharges')
print('la media es',df['TotalCharges'].mean())
print('la mediana es',df['TotalCharges'].median())
```

```python
print('media',df[df['Churn']=='Yes']['tenure'].mean())
print('mediana',df[df['Churn']=='Yes']['tenure'].median())
```

```python
print('media',df[df['Churn']=='No']['tenure'].mean())
print('mediana',df[df['Churn']=='No']['tenure'].median())
```

```python
# Create boxplot
plt.boxplot(df[df['Churn']=='No']['tenure'])
plt.title('Calculo de ternure cuando Churn es no')
plt.xlabel('Group')
plt.ylabel('Value')
plt.show()
```

```python
# Create boxplot
plt.boxplot(df[df['Churn']=='Yes']['tenure'])
plt.title('Calculo de ternure cuando Churn es no')
plt.xlabel('Group')
plt.ylabel('Value')
plt.show()
```

```python
# Create boxplot
plt.boxplot(df[df['Churn']=='Yes']['MonthlyCharges'])
plt.title('Calculo de ternure cuando Churn es no')
plt.xlabel('Group')
plt.ylabel('Value')
plt.boxplot(df[df['Churn']=='No']['MonthlyCharges'])
plt.title('Calculo de ternure cuando Churn es no')
plt.show()
```
