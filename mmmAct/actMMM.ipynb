## Moda, Media y Mediana - EcoBici & AirBnB

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import polars as pl
import io

print("Librerias importadas correctamente")
```

```python
# Leemos los archivos de ecobici y de airbnb
ecobici_fle = pd.read_csv("./2025-04.csv")
airbnb_fle = pl.read_csv('./listings.csv.gz')
```

```python
# Obtenemos información sobre el csv de ecobici
print("Informacion sobre Ecobici:\n")
numfil_bici = ecobici_fle.shape[0] 
numcol_bici = ecobici_fle.shape[1]
print(f"Numero de filas en Ecobici: {numfil_bici}")
print(f"Numero de columnas en Ecobici: {numcol_bici}")

# Obtenemos información sobre el csv de airbnb
print("\nInformacion sobre Airbnb:\n")
numfil_airbnb = airbnb_fle.height
numcol_airbnb = airbnb_fle.width
print(f"Numero de filas en Airbnb: {numfil_airbnb}")
print(f"Numero de columnas en Airbnb: {numcol_airbnb}")
```

```python
# Validamos la cantidad de registros que tenemos por género para ecobici
print(ecobici_fle["Genero_Usuario"].value_counts(dropna=False))
```

```python
# Filtramos y limpiamos la columna de 'genero' paraa ecobici
ecobici_fle['Genero_Usuario'] = ecobici_fle['Genero_Usuario'].where(ecobici_fle["Genero_Usuario"].isin(["M", "F"]), "O")
print(ecobici_fle["Genero_Usuario"].value_counts())
```

```python
# Corregimos valores de edades de ecobici de ser flotantes a ser valaores enteros (no necesitamos valores flotantes en edades)
ecobici_fle["Edad_Usuario"] = ecobici_fle["Edad_Usuario"].astype("Int64")
display(ecobici_fle.head(15))
```

```python
# Cambiar el nombre de la columna "Bici" a "Bici_UUID"
ecobici_fle.rename(columns={"Bici": "Bici_UUID"}, inplace=True)
display(ecobici_fle.head(20))
```

```python
# Ahora, con el uso de 'pandas' vamos a sacar la "moda", "media" y "mediana" de la columna de Edad
edades = "Edad_Usuario"

print(f"Media:\n {ecobici_fle[edades].mean()}")

print(f"\nModa:\n {ecobici_fle[edades].mode().iloc[0]}")

print(f"\nMediana:\n {ecobici_fle[edades].median()}")
```

```python
# Conteo de veces que se repite el mismo UUID de cada bici
conteo = ecobici_fle["Bici_UUID"].value_counts()
print(conteo)
```

```python
# Ahora, para la tabla de airbnb, vamos a hacer una limpieza de la columna de "price"

airbnb_fle = airbnb_fle.drop_nulls("price")
airbnb_fle = airbnb_fle.with_columns(pl.col("price").str.replace_all(r"[\$,]", "").cast(pl.Float64))
print(airbnb_fle.select("price"))
```

```python
# Ahora, seleccionamos las columnas del csv de airbnb de las que queremos obtener la moda, media y mediana

cols_airbnb = ["host_name", "host_location", "host_response_time", "host_response_rate", "host_acceptance_rate", "host_is_superhost", "host_neighbourhood", "host_listings_count", "host_total_listings_count", "accommodates", "bathrooms", "maximum_nights", "has_availability", "number_of_reviews"]

# Filtra las que realmente existen en el DF (evita errores por columnas faltantes)
cols_airbnb = [c for c in cols_airbnb if c in airbnb_fle.columns]

# Helper mini para la moda (funciona con texto o números)
def moda(s: pl.Series):
    vc = s.drop_nulls().value_counts().sort("count", descending=True)
    return vc.select(s.name).to_series()[0] if vc.height > 0 else None

# Calcula media/mediana para numéricas y moda para todas
NUM = {pl.Int8, pl.Int16, pl.Int32, pl.Int64, pl.UInt8, pl.UInt16, pl.UInt32, pl.UInt64, pl.Float32, pl.Float64}

filas = []
for c in cols_airbnb:
    s = airbnb_fle[c]
    if s.dtype in NUM:
        filas.append({"columna": c, "media": s.mean(), "mediana": s.median(), "moda": moda(s)})
    else:
        filas.append({"columna": c, "media": None, "mediana": None, "moda": moda(s)})

resumen = pl.DataFrame(filas)
print(resumen)
```
