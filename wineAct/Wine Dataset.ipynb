```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
wine_data = pd.read_csv(url, delimiter=";")
wine_data.head()
```

```python
wine_data.mean()
```

```python
wine_data.median()
```

```python
wine_data.mode()
```

```python
alcohol_mean = wine_data['alcohol'].mean()
print(f"Mean alcohol content: {alcohol_mean:.2f}")
```

```python
# Plotting
means = wine_data.mean()
plt.figure(figsize=(8,4))
means.plot(kind='bar', color='skyblue')
plt.title('Mean Values of Wine Dataset Features')
plt.ylabel('Mean Value')
plt.xlabel('Features')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
median = wine_data.median()
plt.figure(figsize=(8,4))
median.plot(kind='bar', color='skyblue')
plt.title('Median Values of Wine Dataset Features')
plt.ylabel('Median Value')
plt.xlabel('Features')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
mode = wine_data.mode().iloc[0]
plt.figure(figsize=(8,4))
mode.plot(kind='bar', color='skyblue')
plt.title('Mode Values of Wine Dataset Features')
plt.ylabel('Mode Value')
plt.xlabel('Features')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
# Set up the style for seaborn plot
sns.set(style="whitegrid")

# Selecting the 'alcohol' feature for demonstration
feature = 'alcohol'

plt.figure(figsize=(10, 5))

# Plotting histogram
sns.histplot(wine_data[feature], bins=30, kde=False, color="skyblue")

plt.title(f'Distribución de {feature.capitalize()} en el Dataset de Wine')
plt.xlabel(f'Contenido de {feature.capitalize()}')
plt.ylabel('Frecuencia')

# Highlight the mode
mode_val = wine_data[feature].mode().iloc[0]
plt.axvline(mode_val, color='red', linestyle='dashed', linewidth=2)
min_ylim, max_ylim = plt.ylim()
plt.text(mode_val, max_ylim*0.9, f'Mode: {mode_val:.2f}', horizontalalignment='left', color='red')

plt.tight_layout()
plt.show()
     
```

```python
import seaborn as sns
# Set up the style for seaborn plot
sns.set(style="whitegrid")

# Selecting the 'alcohol' feature for demonstration
feature = 'alcohol'

plt.figure(figsize=(10, 5))

# Plotting histogram
sns.histplot(wine_data[feature], bins=30, kde=False, color="skyblue")

plt.title(f'Distribución de {feature.capitalize()} en el Dataset de Wine')
plt.xlabel(f'Contenido de {feature.capitalize()}')
plt.ylabel('Frecuencia')

mean_value = wine_data[feature].mean()
std_value = wine_data[feature].std()
# Highlight the mode

plt.axvline(mean_value-std_value, color='red', linestyle='dashed', linewidth=2)
plt.text(mean_value-std_value, max_ylim*0.9, f'-1 SD: {mean_value-std_value:.2f}', horizontalalignment='left', color='red')

plt.axvline(mean_value+std_value, color='red', linestyle='dashed', linewidth=2)
plt.text(mean_value+std_value, max_ylim*0.9, f'+ 1 SD: {mean_value+std_value:.2f}', horizontalalignment='left', color='red')

plt.axvline(mean_value, color='green', linestyle='dashed', linewidth=2)
plt.text(mean_value, max_ylim*0.9, f'Media: {mean_value:.2f}', horizontalalignment='left', color='green')
plt.tight_layout()
plt.show()
     
```

```python
import seaborn as sns
sns.set(style="whitegrid")

feature = 'alcohol'

plt.figure(figsize=(10, 5))

sns.histplot(wine_data[feature], bins=30, kde=True, color="skyblue", label="Distribution")

plt.title(f'Distribución de {feature.capitalize()} en el Dataset de Wine')
plt.xlabel(f'Contenido de {feature.capitalize()}')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.legend()
plt.show()
     
```

```python

# Compute quartiles and IQR (using percentile function)
Q1 = np.percentile(wine_data['alcohol'], 25)
Q2 = np.percentile(wine_data['alcohol'], 50)
Q3 = np.percentile(wine_data['alcohol'], 75)
IQR = Q3 - Q1

print(f"Q1 (25th Percentile): {Q1:.4f}")
print(f"Q2 (Median / 50th Percentile): {Q2:.4f}")
print(f"Q3 (75th Percentile): {Q3:.4f}")
print(f"IQR (Interquartile Range): {IQR:.4f}")

# Visualization using a boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=wine_data['alcohol'], color="lightblue")
plt.title("Boxplot del Contenido de Alcohol  en el dataset de Wine")
plt.xlabel('Porcentaje de alcohol (%)')

# Displaying the quartile values on the plot
plt.annotate(f'Q1: {Q1:.2f}', xy=(Q1, -0.4), xytext=(Q1, -0.45), arrowprops=dict(arrowstyle='->', color='red'), color='red')
plt.annotate(f'Q2: {Q2:.2f}', xy=(Q2, -0.35), xytext=(Q2, -0.4), arrowprops=dict(arrowstyle='->', color='green'), color='green')
plt.annotate(f'Q3: {Q3:.2f}', xy=(Q3, -0.4), xytext=(Q3, -0.45), arrowprops=dict(arrowstyle='->', color='blue'), color='blue')

plt.tight_layout()
plt.show()
```
